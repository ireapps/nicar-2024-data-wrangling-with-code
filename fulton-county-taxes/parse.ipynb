```python
# get a list of column lists from the parsing rules dict
cols = [parsing_rules[x]['usecols'] for x in parsing_rules]

# flatten that list of lists and grab just the unique values
cols_flat = list(set([item for sublist in cols for item in sublist]))

# print results, sorted
print(sorted(cols_flat))
```

```python
# map current column names to canonical names
column_map = {
    'BID AMOUNT': 'bid_amount',
    'BID AMOUNT / DEED FACE': 'bid_amount',
    'FIFA ADDRESS': 'address',
    'MIN BID': 'min_bid',
    'MINIMUM BID': 'min_bid',
    'MINIMUM BID': 'min_bid',
    'OWNER': 'owner_name',
    'OWNER NAME': 'owner_name',
    'PARCEL': 'parcel_id',
    'PARCEL ID': 'parcel_id',
    'PURCHASER': 'purchaser_name',
    'PURCHASER NAME': 'purchaser_name',
    'PURCHASER NAME ': 'purchaser_name',
    'SITUS': 'address'
}
```

```python
# create a new empty data frame
df = pd.DataFrame()
```

```python
# loop over each filename + parsing rules
for filename in parsing_rules:
    
    # extract the year and month in this complicated list comprehension
    year, month = [int(x) for x in filename.split('/')[-1].rstrip('.xls').split('_')]
    
    # create a data frame by reading in the file and applying the parsing rules
    # https://towardsdatascience.com/unpacking-operators-in-python-306ae44cd480
    this_df = pd.read_excel(filename, **parsing_rules[filename])
    
    this_df = this_df[sorted(this_df.columns)]
    
    cols_canon = [column_map.get(x) for x in this_df.columns]
    
    # rename the columns in the data frame
    this_df.columns = cols_canon
    
    # add month, year and filename values
    this_df['year'] = year
    this_df['month'] = month
    this_df['filename'] = filename.split('/')[-1]
    
    # add it to the df created above
    df = pd.concat([df, this_df])
```

```python
# replace NaN vals with empty strings across the board
df = df.fillna('')
```

```python
# fix parcel ID formatting by removing all nonalphanumeric characters from the string
# https://realpython.com/python-lambda/
df['parcel_id'] = df.apply(lambda row: ''.join([x for x in row['parcel_id'] if x.isalnum()]), axis=1)
```
